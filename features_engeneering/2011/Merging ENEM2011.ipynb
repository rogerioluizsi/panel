```python
import pandas as pd
```

```python
#ENEM 2011 was in SAS and TXT data type. I split them to open in low hardware SAS system and converted to CSV. 
#Now, I merging the splitted parts.

e1=pd.read_csv("ENEM_2011_1.csv")
e2a=pd.read_csv("ENEM_2011_2A.csv")
e2b1=pd.read_csv("ENEM_2011_2b1.csv")
e2b2=pd.read_csv("ENEM_2011_2b2.csv")
e3=pd.read_csv("ENEM_2011_3.csv")
```

```python
e2b = pd.merge(e2b1, e2b2, left_on='NU_INSCRICAO', right_on='NU_INSCRICAO', how='inner')
e2 = pd.merge(e2b, e2a, left_on='NU_INSCRICAO', right_on='NU_INSCRICAO', how='inner')
```

```python
print (e1.info())
print (e2.info())
print(e3.info())
```

### The avalilable hardware could not process this datasets. Let 's get downcasting them

```python
#dowcasting numeric feature to improve memory usage.
num_cols = e1.select_dtypes(include =['int']).columns
e1[num_cols]= e1[num_cols].apply(pd.to_numeric,downcast='unsigned')
```

```python
#downside float features to improve memory usage
flo_cols = e1.select_dtypes(include =['float']).columns
e1[flo_cols]= e1[flo_cols].apply(pd.to_numeric,downcast='float')
```

```python
num_cols = e2.select_dtypes(include =['int']).columns
e2[num_cols]= e2[num_cols].apply(pd.to_numeric,downcast='unsigned')
```

```python
#downside float features to improve memory usage
flo_cols = e2.select_dtypes(include =['float']).columns
e2[flo_cols]= e2[flo_cols].apply(pd.to_numeric,downcast='float')
```

```python
num_cols = e3.select_dtypes(include =['int']).columns
e3[num_cols]= e3[num_cols].apply(pd.to_numeric,downcast='unsigned')
```

```python
#downside float features to improve memory usage
flo_cols = e3.select_dtypes(include =['float']).columns
e3[flo_cols]= e3[flo_cols].apply(pd.to_numeric,downcast='float')
```

```python
#merging the rest of the data parts
e12 = pd.merge(e1, e2, left_on='NU_INSCRICAO', right_on='NU_INSCRICAO', how='inner')
e = pd.merge(e12, e3, left_on='NU_INSCRICAO', right_on='NU_INSCRICAO', how='inner')
```

```python
#checking the memeroy use for each dataset now!
print (e1.info())
print (e2.info())
print(e3.info())
```

```python
#for the new dataset, the memory use seems half as much as before
e.info(memory_usage='deep')
```
